```python
# Import libraries

import matplotlib.pylab as plt
import cv2
import os
import numpy as np
```

```python
# Read cluster label file

fp = open('reid_final_clusters.csv', 'r')
next(fp)
image_path = []
labels = []
for line in fp:
  path, label = line.strip().split(",")
  image_path.append(path)
  labels.append(int(label))
labels = np.array(labels)
```

```python
# Plot cluster based on cluster id

def plot_clusters(cluster_id):
  cluster = []
  occurence = np.where(labels == cluster_id)[0]
  for i in occurence:
    image=plt.imread('../../dataset/'+image_path[i])
    cluster.append(image)
  cluster=np.array(cluster)

  f, axs = plt.subplots(1, len(occurence), figsize=(5, 2*len(occurence)))
  for j in range(len(occurence)):
      axs[j].imshow(cluster[j], cmap='binary')
      axs[j].axis('off')
```

```python
plot_clusters(97)
```

```python
plot_clusters(255)
```

```python
plot_clusters(356)
```
